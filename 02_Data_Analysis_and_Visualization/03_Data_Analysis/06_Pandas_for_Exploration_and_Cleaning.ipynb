# Pandas for Exploration and Cleaning

```python
# Impor the pandas package:
import pandas as pd

# Extract the new CSV data:
df = pd.read_csv("Archivos_CSV_Excel_y_JSON/06_sales_data_new.csv")
print(df.head())
```

```python
# show the Info of the DataFrame:
print(df.info())
```

```python
# Count the null values:
print(df.isnull().sum())
```

Remove the Irrelevant Column

```python
# Remove the irrelevant column:
df.drop("Unnamed: 0", axis=1, inplace=True)
```

Update the Date

```python
df['date'] = pd.to_datetime(df['date'])
```

Agregate the DataFrame

```python
sales_by_product = df.groupby('product')['sales'].sum()
print(sales_by_product)
```

Create Profit Column

```python
# Debug and eradicate any strings or excesive floats
df = df.dropna()
df['price'] = [int (float (x)) for x in df['price']]
df['cost'] = [int (float (x)) for x in df['cost']]
print(df.describe())
```
