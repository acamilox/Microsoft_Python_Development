# Transforming Data with Pandas

```python
# Import the neccesary packages
import pandas as pd

# Extract the csv file for this assay:
df = pd.read_csv('Archivos_CSV_Excel_y_JSON/05_sales_data.csv')
print(df.head())
```

Sort the DataFrame

```python
df_sorted = df.sort_values(by='Sales', ascending=False)
print(df_sorted.head())
```

Group the DataFrame

```python
df_grouped = df.groupby('Region')['Sales'].sum()
print(df_grouped)
```

Agregate the DataFrame

```python
mean_sales = df['Sales'].mean()
print(mean_sales)
```

Disccount the Dataframe

```python
def calculate_discount(sales):
    return sales * 0.9

df['Discounted_Sales'] = df['Sales'].apply(calculate_discount)
print(df.head())
```

Pivot the DataFrame

```python
df_pivot = df.pivot_table(index='Region', columns='Product', values='Sales')
print(df_pivot)
```
