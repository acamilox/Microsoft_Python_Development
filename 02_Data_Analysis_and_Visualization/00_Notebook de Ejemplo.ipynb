# Uso de la función %matplotlib inline

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 3*np.pi, 500)
plt.plot(x, np.sin(x**2))
plt.title("A simple chirp")
```

```python
import random
import pandas as pd
import matplotlib.pyplot as plt

# Define a list of 10 customer names
customer_names = ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace',
                  'Hannah', 'Ian', 'Jane']

# Define a list of sample purchase items
purchase_items = ['Laptop', 'Smartphone', 'Tablet', 'Headphones', 'Camera',
                  'Smartwatch', 'Printer', 'Monitor', 'Keyboard', 'Mouse']

# Generate a dataset with 100 random purchases
expanded_data = {
    'Customer Name': [random.choice(customer_names) for _ in range(100)],
    'Purchase Item': [random.choice(purchase_items) for _ in range(100)],
    'Purchase Cost': [round(random.uniform(20.0, 2000.0), 2) for _ in range(100)]
}

# Convert to a DataFrame
df = pd.DataFrame(expanded_data)

# Display the dataset
print(df.head())
```

# Visualizar los datos con un gráfico de barras

```python
# Plot a bar chart
plt.figure(figsize=(15, 5))
plt.bar(df['Customer Name'], df['Purchase Cost'], color='green')
plt.xlabel('Customer Name')
plt.ylabel('Purchase Cost ($)')
plt.title('Customer Purchases')
plt.xticks(rotation=45)
plt.show()
```
